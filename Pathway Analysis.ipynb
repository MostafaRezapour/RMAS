```python
import pandas as pd


# Function to extract sets of genes
def extract_gene_sets(df_list):
    return [set(df['Gene Symbol'].unique()) for df in df_list]

# Extract sets of upregulated genes for easier comparison
upregulated_genes_sets_24hrs = extract_gene_sets(upregulated_24hrs)
upregulated_genes_sets_72hrs = extract_gene_sets(upregulated_72hrs)

# Virus labels for clarity in output
viruses = ['IAV', 'MPV', 'PIV3']

# Compare gene expression profiles from 24 to 72 hours for each virus
for i, virus in enumerate(viruses):
    consistently_upregulated = upregulated_genes_sets_24hrs[i] & upregulated_genes_sets_72hrs[i]
    newly_upregulated = upregulated_genes_sets_72hrs[i] - upregulated_genes_sets_24hrs[i]
    no_longer_upregulated = upregulated_genes_sets_24hrs[i] - upregulated_genes_sets_72hrs[i]
    
    print(f"\n{virus} - Consistently Upregulated Genes: {', '.join(consistently_upregulated)}")
    print(f"{virus} - Newly Upregulated Genes at 72 Hours: {', '.join(newly_upregulated)}")
    print(f"{virus} - No Longer Upregulated Genes at 72 Hours: {', '.join(no_longer_upregulated)}")
```
